## Cell 3 of 5 — Define the 5 Consumer Personas

What this does: Creates 5 fictional but realistic consumer profiles.

Each persona has demographics and three psychological traits from the Twin-2K-500 framework:
- **Self-monitoring:** sensitivity to social cues and concern with how one appears to others
- **Need for uniqueness:** desire to differentiate oneself from the mainstream
- **Minimalism:** preference for simplicity and skepticism toward conspicuous consumption

These three traits were selected because they are directly relevant to how consumers process influencer wealth displays:
- High self-monitors are attuned to status signals and more likely to be influenced by wealth cues
- High need for uniqueness predicts rejection of influencers who feel commercially co-opted
- High minimalism predicts criticism of conspicuous wealth display as inauthentic or excessive

**Run this cell now.**

```python
personas = [
    {
        "name": "Participant 1 - Zoe",
        "profile": (
            "You are Zoe, a 22-year-old aspiring content creator with 8,000 TikTok followers "
            "and 5,000 Instagram followers, living in Los Angeles. "
            "You earn $28,000/year working part-time at a boutique while building your own presence. "
            "You have HIGH self-monitoring, LOW need for uniqueness, and LOW minimalism. "
            "You actively follow mega-influencers and find their lifestyle aspirational and motivating. "
            "Wealth displays feel like proof that the influencer career path works, not a reason to criticize. "
            "You follow accounts on both TikTok and Instagram daily. "
            "Answer in first person as Zoe. Keep answers to 3-5 sentences per question."
        )
    },
    {
        "name": "Participant 2 - Marcus",
        "profile": (
            "You are Marcus, a 34-year-old high school art teacher earning $52,000/year in Portland, Oregon. "
            "You use Instagram occasionally but find TikTok overwhelming. "
            "You have LOW self-monitoring, HIGH need for uniqueness, and HIGH minimalism. "
            "You find influencer culture shallow and wealth displays particularly off-putting. "
            "You are sensitive to the difference between genuine success and performative richness. "
            "When an influencer flaunts expensive items you immediately question their authenticity and motives. "
            "Answer in first person as Marcus. Keep answers to 3-5 sentences per question."
        )
    },
    {
        "name": "Participant 3 - Priya",
        "profile": (
            "You are Priya, a 28-year-old marketing analyst earning $67,000/year in Chicago. "
            "You use both TikTok and Instagram daily and follow a mix of lifestyle and finance influencers. "
            "You have MODERATE self-monitoring, MODERATE need for uniqueness, and MODERATE minimalism. "
            "Your reaction to influencer wealth depends heavily on the perceived source: "
            "brand deals and business income feel legitimate, while debt-funded or family-money displays feel dishonest. "
            "You have unfollowed influencers whose wealth felt unearned or misrepresented. "
            "Answer in first person as Priya. Keep answers to 3-5 sentences per question."
        )
    },
    {
        "name": "Participant 4 - Derek",
        "profile": (
            "You are Derek, a 41-year-old financial analyst earning $110,000/year in New York City. "
            "You use Instagram but not TikTok. You follow a small number of finance and business accounts. "
            "You have LOW self-monitoring, LOW need for uniqueness, and LOW minimalism. "
            "You evaluate influencer wealth displays analytically: you can estimate roughly what a given "
            "follower count generates in brand deal income and are skeptical of displays inconsistent with plausible earnings. "
            "You are unbothered by wealth itself but sharply critical of financial dishonesty or misleading self-presentation. "
            "Answer in first person as Derek. Keep answers to 3-5 sentences per question."
        )
    },
    {
        "name": "Participant 5 - Aaliyah",
        "profile": (
            "You are Aaliyah, a 26-year-old sociology graduate student earning $22,000/year in Atlanta. "
            "You are an active TikTok and Instagram user who engages critically with influencer culture. "
            "You have HIGH self-monitoring, HIGH need for uniqueness, and HIGH minimalism. "
            "You are deeply attuned to social signals but actively reject mainstream status cues. "
            "This creates real tension: you notice and analyze wealth displays closely, "
            "but they make you feel critical and sometimes angry rather than aspirational. "
            "You are especially bothered by debt-funded or family-money wealth presented as self-made success. "
            "Answer in first person as Aaliyah. Keep answers to 3-5 sentences per question."
        )
    }
]

print(f"{len(personas)} consumer personas defined. Ready for Cell 4.")
for p in personas:
    print(f"  - {p['name']}")
print()
print("Trait summary:")
print("  Zoe:     HIGH self-monitoring | LOW uniqueness  | LOW minimalism  -> Aspirational")
print("  Marcus:  LOW self-monitoring  | HIGH uniqueness | HIGH minimalism -> Critical")
print("  Priya:   MODERATE all traits                                      -> Source-sensitive")
print("  Derek:   LOW self-monitoring  | LOW uniqueness  | LOW minimalism  -> Analytical")
print("  Aaliyah: HIGH self-monitoring | HIGH uniqueness | HIGH minimalism -> Conflicted critic")
```

---
## Cell 4 of 5 — Run the Interviews

What this does: Sends each persona to GPT-4 with 3 interview questions and collects responses.

The 3 questions focus on TikTok and Instagram contexts:
1. How do you react when an influencer you follow posts content showing expensive purchases or a luxurious lifestyle?
2. Does it matter to you WHERE an influencer's wealth comes from — brand deals, family money, debt, or a secondary job?
3. Have you ever unfollowed, criticized, or changed your behavior toward an influencer because of how they displayed their wealth?

What you will see: A status update as each interview completes. Takes about 30-60 seconds.

If you see an error: The most common cause is an incorrect API key. Go back to Cell 2 and re-run it.

**Run this cell now.**

```python
interview_questions = [
    "How do you react when an influencer you follow on TikTok or Instagram posts content "
    "showing expensive purchases, luxury travel, or a lavish lifestyle?",

    "Does it matter to you WHERE an influencer's wealth comes from? "
    "For example: does it change how you feel if the money comes from brand deals, "
    "family wealth, debt, or a separate job outside of social media?",

    "Have you ever unfollowed, criticized, or changed your purchasing behavior "
    "toward an influencer specifically because of how they displayed their wealth? "
    "Tell me what happened."
]

def run_interview(persona, questions):
    questions_text = "\n".join([f"{i+1}. {q}" for i, q in enumerate(questions)])
    messages = [
        {"role": "system", "content": persona["profile"]},
        {"role": "user",   "content": f"Please answer each question below. Label answers Q1, Q2, Q3.\n\n{questions_text}"}
    ]
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=messages,
        temperature=0.7
    )
    return response.choices[0].message.content

print("Running interviews...\n")
results = {}
for persona in personas:
    print(f"  Interviewing {persona['name']}...", end=" ")
    results[persona["name"]] = run_interview(persona, interview_questions)
    print("Done")

print("\nAll 5 interviews complete. Ready for Cell 5.")
```

---
## Cell 5 of 5 — Display and Compare the Responses

What this does: Prints all 5 interview responses so you can compare them.

What to look for:
- Does Zoe (high self-monitoring, low minimalism) respond differently from Marcus (low self-monitoring, high minimalism) to the same wealth display?
- Does Priya's source-sensitivity show up clearly — does she distinguish brand deals from debt-funded wealth?
- Aaliyah has the theoretically interesting profile: high self-monitoring means she NOTICES wealth signals, but high need for uniqueness and high minimalism mean she REJECTS them. Does that tension appear in her answer?
- How does Derek's financial literacy shape his reaction compared to the others?

**Run this cell now.**

```python
print("=" * 70)
print("SYNTHETIC INTERVIEW RESULTS: INFLUENCER WEALTH PERCEPTION")
print("Consumer Digital Twin Analysis - Week 6")
print("=" * 70)
print("\nQUESTIONS ASKED:")
for i, q in enumerate(interview_questions):
    print(f"  Q{i+1}: {q}")
print("\n" + "=" * 70)

for name, response in results.items():
    print(f"\n{'-' * 70}")
    print(f"PARTICIPANT: {name}")
    print("-" * 70)
    print(response)

print("\n" + "=" * 70)
print("END OF INTERVIEWS")
print("=" * 70)
print("""
REFLECTION QUESTIONS FOR CLASS DISCUSSION:

1. Priya is the only persona explicitly framed as source-sensitive.
   Did the other personas spontaneously distinguish between wealth sources
   (brand deals vs. debt vs. family money) without being prompted?

2. Aaliyah has the conflicted profile: she notices wealth signals
   (high self-monitoring) but rejects them (high need for uniqueness,
   high minimalism). This is the theoretically interesting case for
   the research. Does her response reflect that internal conflict?

3. The research uses Reddit data to study real consumer reactions.
   Which of these five personas seems most likely to post a Reddit
   comment criticizing an influencer's wealth display?
   Does that prediction match their trait profile?

4. What would change if we replaced these hand-written personas with
   real Twin-2K-500 profiles retrieved via RAG -- specifically selecting
   participants based on their score_selfmonitoring,
   score_needforuniqueness, and score_minimalism scores?
""")
```
