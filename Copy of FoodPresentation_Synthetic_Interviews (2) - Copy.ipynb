## Cell 3 of 5 — Define the 5 Consumer Personas

What this does: Creates 5 fictional but realistic consumer profiles for restaurant diners.

Each persona varies on three psychological traits from the Twin-2K-500 dataset framework:

- **Self-monitoring:** How much the person cares about appearance and social impressions of food
- **Need for uniqueness:** Whether they value rare, artisanal, or visually distinctive experiences
- **Present bias:** Whether they prioritize immediate sensory pleasure over restraint

These traits were selected because they are most theoretically relevant to food presentation appeal.

**Run this cell now.**

```python
personas = [
    {
        "name": "Participant 1 - Sophia",
        "profile": (
            "You are Sophia, a 29-year-old food blogger and Instagram influencer in New York City. "
            "You dine out frequently at upscale restaurants and photograph every meal before eating. "
            "You have VERY HIGH self-monitoring (you are acutely aware of how food looks to others), "
            "VERY HIGH need for uniqueness (you seek out rare and visually distinctive dishes), "
            "and HIGH present bias (you live for immediate sensory pleasure). "
            "You are deeply attuned to food aesthetics and presentation style. "
            "Answer in first person as Sophia. Keep answers to 3-5 sentences per question."
        )
    },
    {
        "name": "Participant 2 - Frank",
        "profile": (
            "You are Frank, a 58-year-old retired construction foreman in rural Pennsylvania. "
            "You eat at diners and family restaurants. You value large portions and hearty meals. "
            "You have LOW self-monitoring (you do not care how food looks to others), "
            "LOW need for uniqueness (you prefer familiar comfort food over anything artisanal), "
            "and LOW present bias (you are practical and value substance over appearance). "
            "You are skeptical of trendy small-plate restaurant concepts. "
            "Answer in first person as Frank. Keep answers to 3-5 sentences per question."
        )
    },
    {
        "name": "Participant 3 - Anika",
        "profile": (
            "You are Anika, a 35-year-old pediatrician in Chicago who dines out once or twice a week. "
            "You are health-conscious and enjoy quality food but are not obsessed with aesthetics. "
            "You have MODERATE self-monitoring, MODERATE need for uniqueness, "
            "and MODERATE present bias. "
            "You appreciate beautiful food presentation but also want to feel satisfied after a meal. "
            "Answer in first person as Anika. Keep answers to 3-5 sentences per question."
        )
    },
    {
        "name": "Participant 4 - Marcus",
        "profile": (
            "You are Marcus, a 26-year-old graduate student in food science at UC Davis. "
            "You are analytically minded about food and interested in culinary innovation. "
            "You have HIGH self-monitoring (you notice and analyze food presentation carefully), "
            "HIGH need for uniqueness (you actively seek out novel dining experiences), "
            "and LOW present bias (you think carefully before indulging). "
            "You find small tasting portions intellectually interesting as a format. "
            "Answer in first person as Marcus. Keep answers to 3-5 sentences per question."
        )
    },
    {
        "name": "Participant 5 - Linda",
        "profile": (
            "You are Linda, a 47-year-old office manager in suburban Ohio who dines out on special occasions. "
            "You are budget-conscious and feel anxious about leaving a restaurant hungry. "
            "You have LOW self-monitoring (you care little about how food looks), "
            "LOW need for uniqueness (you prefer familiar dishes done well), "
            "and HIGH present bias (you want immediate satisfaction and fullness). "
            "You are suspicious of small portions and feel they signal poor value for money. "
            "Answer in first person as Linda. Keep answers to 3-5 sentences per question."
        )
    }
]

print(f"{len(personas)} consumer personas defined. Ready for Cell 4.")
for p in personas:
    print(f"  - {p['name']}")
```

---
## Cell 4 of 5 — Run the Interviews

What this does: Sends each persona to GPT-4 with 3 restaurant-focused interview questions.

The 3 questions:
1. Imagine you are in a restaurant and you see the people at a table next to you have been served a large dish of food to share from while others have received a bunch of small dishes to take for themselves, which one makes you more hungry? And why?
2. Does the size of a food presentation affect how much you want to eat it? Does a smaller, more elegant presentation make the food more or less appealing to you?
3. Think about a time when you saw a restaurant or food advertisement that made you hungry. How did the presentation elements (such as sizing) affect your overall experience and satisfaction?

What you will see: A status update as each interview completes. Takes about 30-60 seconds.

If you see an error: Go back to Cell 2 and re-run it with your API key.

**Run this cell now.**

```python
interview_questions = [
    "Imagine you are in a restaurant and you see the people at a table next to you have been served a large dish of food to share from while others have received a bunch of small dishes to take for themselves, which one makes you more hungry? And why?",
    "Does the size of a food presentation affect how much you want to eat it? Does a smaller, more elegant presentation make the food more or less appealing to you?",
    "Think about a time when you saw a restaurant or food advertisement that made you hungry. How did the presentation elements (such as sizing) affect your overall experience and satisfaction?"
]

def run_interview(persona, questions):
    questions_text = "\n".join([f"{i+1}. {q}" for i, q in enumerate(questions)])
    messages = [
        {"role": "system", "content": persona["profile"]},
        {"role": "user", "content": f"Please answer each question below. Label answers Q1, Q2, Q3.\n\n{questions_text}"}
    ]
    response = client.chat.completions.create(
        model="gpt-4",
        messages=messages,
        temperature=0.7
    )
    return response.choices[0].message.content

print("Running interviews...\n")
results = {}
for persona in personas:
    print(f"  Interviewing {persona['name']}...", end=" ")
    results[persona["name"]] = run_interview(persona, interview_questions)
    print("Done")

print("\nAll 5 interviews complete. Ready for Cell 5.")
```

---
## Cell 5 of 5 — Display and Compare the Responses

What this does: Prints all 5 interview responses for comparison.

What to look for:
- Does Sophia (very high self-monitoring, high need for uniqueness) embrace small presentations?
- Does Frank (low self-monitoring, low need for uniqueness) reject them?
- Does Marcus (high uniqueness, low present bias) engage analytically rather than emotionally?
- Do Linda and Frank converge even though their reasons may differ?

**Run this cell now.**

```python
print("=" * 70)
print("SYNTHETIC INTERVIEW RESULTS: FOOD SIZE PRESENTATION")
print("Consumer Digital Twin Analysis - Week 6")
print("=" * 70)
print("\nQUESTIONS ASKED:")
for i, q in enumerate(interview_questions):
    print(f"  Q{i+1}: {q}")
print("\n" + "=" * 70)

for name, response in results.items():
    print(f"\n{chr(45) * 70}")
    print(f"PARTICIPANT: {name}")
    print(chr(45) * 70)
    print(response)

print("\n" + "=" * 70)
print("END OF INTERVIEWS")
print("=" * 70)
print("""
REFLECTION QUESTIONS FOR CLASS DISCUSSION:

1. Which persona found small food presentations most appealing?
   Does self-monitoring or need for uniqueness seem to drive that more?

2. Frank and Linda both have low self-monitoring and low need for uniqueness.
   Do their responses differ? What explains any differences you see?

3. Marcus is analytical about food. Does his response feel different in tone
   from Sophia, who is emotional and aesthetic? What does that tell us
   about the same preference (liking small portions) can come from
   different psychological pathways?

4. How could a restaurant use these profiles to design menus or
   plate presentations for different customer segments?

5. What Twin-2K-500 variables would you want to use to ground
   these personas in real participant data via RAG?
""")
```
